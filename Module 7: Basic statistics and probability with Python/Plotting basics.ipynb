```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
%matplotlib inline
```

## Line Plots

```python
# Generate 100 values 
x = np.random.randint(1, 10, 100)
y = x + 5

plt.plot(x)
plt.plot(y)
plt.legend(['x', 'y'])
plt.ylabel('Values')
plt.title('Random Series Shift')

plt.show()
```

In this line plot, x and y are parallel to each other. The y value has been shifted up 5 compared to the x value. 

## Scatterplot 

```python
# Generate 20 random vlaues. 

x = np.random.randint(1, 100, 20)
y = np.random.randint(1, 100, 20)
y2 = np.random.randint(1, 100, 20)
x2 = np.random.randint(1, 100, 20)
plt.figure(figsize = (6, 4))

plt.scatter(x, y, label = "first", c = "b", s = 250, marker = "^")
plt.scatter(x2, y2, label = "second", c = "orange", s = 250, marker = "s")

plt.xlabel("X", fontsize = 15) ; plt.ylabel("Y", fontsize = 15)
plt.xticks(size = 12) ; plt.yticks(size = 12)
plt.title("Random Scatter Plot", fontsize = 18)
plt.legend(fontsize = 12)
plt.show();
```

There doesn't appear to be a strong coorelation between first and second points. Both first and second points are generally distributed throughout the scatter plot. 

## Histogram

```python
# Generate 1000 random values. 
x = np.random.randint(1, 10, 1000)

plt.hist(x, normed=True, color='blue')
plt.title('Histogram')
plt.xlabel('Random Values')

plt.hist(x)
plt.show()
```

This histogram is not normalized as it does not resemble a bell shape curve. Although the mode seems to be around 2, this histogram appears to be slightly skewed to the right.  

## Boxplot 

```python
np.random.seed(80)

x = np.random.normal(1, 10, 500)

plt.boxplot(x)
plt.show()
```

It appears that the median is about zero. The 75th precentile seems to be skewed towards 30 since there are a couple of outliners near 30.
