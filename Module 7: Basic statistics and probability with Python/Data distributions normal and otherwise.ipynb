```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1 Logisitics

loc, scale = 10, 1
l = np.random.logistic(loc, scale, 100)

mean = np.mean(l)
sd = np.std(l)

plt.hist(l, bins=50, color='c')
plt.axvline(x=mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(x=mean+sd, color='b', linestyle='dashed', linewidth=2)
plt.axvline(x=mean-sd, color='b', linestyle='dashed', linewidth=2)
plt.show()

print('The mean appears to be around 9.8 and the distribution is skewed to the right.')
```

```python
# 2 Triangular 

h = np.random.triangular(-3, 0, 8, 100)

mean = np.mean(h)
sd = np.std(h)

plt.hist(h, bins=50, color='c')
plt.axvline(x=mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(x=mean+sd, color='b', linestyle='dashed', linewidth=2)
plt.axvline(x=mean-sd, color='b', linestyle='dashed', linewidth=2)
plt.show()

print('The mean appears to be around 1.99 and the distribution is skewed to the left.')
```

```python
# 3 Vonmises

mu, kappa = 0.0, 4.0 # mean and dispersion
s = np.random.vonmises(mu, kappa, 100)

mean = np.mean(s)
sd = np.std(s)

plt.hist(s, bins=50, density=True, color='c')
plt.axvline(x=mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(x=mean+sd, color='b', linestyle='dashed', linewidth=2)
plt.axvline(x=mean-sd, color='b', linestyle='dashed', linewidth=2)
plt.show()

print('The mean appears to be around 0.0 and the data is close to being normally distributed.')
```

```python
# 4 Uniform

s = np.random.uniform(-1, 0, 100)

mean = np.mean(s)
sd = np.std(s)

plt.hist(s, bins=25, density=True, color='c')
plt.axvline(x=mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(x=mean+sd, color='b', linestyle='dashed', linewidth=2)
plt.axvline(x=mean-sd, color='b', linestyle='dashed', linewidth=2)
plt.show()

print('The mean appears to be around -0.45 and the distribution is slightly skewed to the right.')
```

```python
# 5 Wald

h = np.random.wald(3, 2, 100)

mean = np.mean(h)
sd = np.std(h)

plt.hist(h, bins=25, density=True, color='c')
plt.axvline(x=mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(x=mean+sd, color='b', linestyle='dashed', linewidth=2)
plt.axvline(x=mean-sd, color='b', linestyle='dashed', linewidth=2)
plt.show()

print('The mean appears to be around 2.6 and the distribution is skewed to the left.')
```

```python
# 6 Laplace

loc, scale = 0., 1.
s = np.random.laplace(loc, scale, 100)

mean = np.mean(s)
sd = np.std(s)

plt.hist(s, bins=35, density=True, color='c')
plt.axvline(x=mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(x=mean+sd, color='b', linestyle='dashed', linewidth=2)
plt.axvline(x=mean-sd, color='b', linestyle='dashed', linewidth=2)
plt.show()

print('The mean appears to be around 0.01 and the distribution is slightly skewed to the right.')
```

```python
# Additionally 

x = np.random.normal(5, 0.5, 100)
y = np.random.normal(10, 1, 100)
z = x+y 

mean= np.mean(z)
sd = np.std(z)

plt.hist(z, color='c')
plt.axvline(x=mean,color='g',linestyle='solid', linewidth=2)
plt.axvline(x=mean+sd, color='g', linestyle='solid', linewidth=2)
plt.axvline(x=mean-sd, color='g', linestyle='solid', linewidth=2)
plt.show()
print('The mean is close to 14.9 and it appears to be normally distributed.')
```
