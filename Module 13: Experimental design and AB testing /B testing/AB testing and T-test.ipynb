```python
# Create dataframe for t-values

df = pd.DataFrame({'y1':[5,1090,45],
                  'y2':[8,999,40],
                  's1':[1,400,45],
                  's2':[3,30,40],
                  'N1':[200,900,2000],
                  'N2':[500,100,2000]})
df.head()
```

```python
# Create function to calculate t-values

t_values = []
for index, row in df.iterrows():
    t = (row['y1'] - row['y2']) / ((row['s1']**2 /row['N1'] + row['s2']**2 / row['N2'])**0.5)
    t_values.append(t)
    
print(t_values)
```
